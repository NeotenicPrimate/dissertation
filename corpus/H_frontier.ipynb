```python
import networkx as nx
import random
import matplotlib.pyplot as plt
```

```python
T = 10

c1 = nx.gnp_random_graph(10, 0.5)
c2 = nx.gnp_random_graph(10, 0.5)

c1 = nx.relabel_nodes(c1, dict(zip(c1.nodes, range(0, len(c1)))))
c2 = nx.relabel_nodes(c2, dict(zip(c2.nodes, range(len(c1), len(c1)+len(c2)))))

G = nx.Graph()
G = nx.disjoint_union(G, c1)
G = nx.disjoint_union(G, c2)

for t in range(T):

    nx.draw(G)
    plt.show()

    n1 = random.choice(list(c1.nodes()))
    n2 = random.choice(list(c2.nodes()))

    print(n1, n2)
    
    G.add_edge(n1, n2)
```

```python
T = 10

G = nx.gnp_random_graph(20, 0.5)


c1 = random.choices(list(G.nodes), k=int(len(G)/2))
c2 = [n for n in G.nodes if n not in c1]

nx.draw(G)

for t in range(T):

    nx.draw(G)
    plt.show()

    n1 = random.choice(c1)
    n2 = random.choice(c2)

    print(n1, n2)

    if G.has_edge(n1, n2):
        G.remove_edge(n1, n2)
```
