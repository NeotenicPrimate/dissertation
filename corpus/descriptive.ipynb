## Centralization

```python
for (field_name, corpus) in corpora.data.items():

    G = corpus['citation_graph']
    centralities = nx.in_degree_centrality(G).values()
    max_centrality = max(centralities)
    d = sum(c - max_centrality for c in centralities)

    G_star = nx.star_graph(len(G)-1).to_directed()
    for (u, v) in list(G_star.edges()):
        if u == 0: G_star.remove_edge(u, v)

    centralities_star = nx.in_degree_centrality(G_star).values()
    max_centrality_star = max(centralities_star)
    d_star = sum(c - max_centrality_star for c in centralities_star)

    centralization = d / d_star
    print(f'''
        {field_name.capitalize()} 
        Centralization: {centralization}
    ''')
```

## Isolates, Loops, Cycles

```python
for (field_name, corpus) in corpora.data.items():

    G = corpus['citation_graph']

    orgs_isolates = G.remove_nodes_from(nx.isolates(G))
    orgs_loop = G.remove_edges_from(nx.selfloop_edges(G))
    orgs_cycles = G.remove_edges_from([(c[-1], c[0]) for c in nx.simple_cycles(G)])

    
    print(f'''
        {field_name.capitalize()} 
        Isolates: {orgs_isolates}
        Loops:    {orgs_loop}
        Cycles:   {orgs_cycles}
    
    ''')
```

## Average Degree

```python
for (field_name, corpus) in corpora.data.items():

    G = corpus['citation_graph']

    n_strongly_connected_components = len(list(nx.strongly_connected_components(G)))
    n_weakly_connected_components = len(list(nx.weakly_connected_components(G)))

    in_degree_map = dict(G.in_degree())
    in_degree = in_degree_map.values()
    avg_in_degree = sum(in_degree) / len(in_degree)
    
    out_degree_map = dict(G.out_degree())
    out_degree = out_degree_map.values()
    avg_out_degree = sum(out_degree) / len(out_degree)

    print(f'''
        {field_name.capitalize()} 
        Avg In Degree: {avg_in_degree}
        Avg Out Degree: {avg_out_degree}
    ''')
```

## Logarithmically growing geodesic length

```python
for (field_name, corpus) in corpora.data.items():

    df = corpus['citation_df']
    G = corpus['citation_graph']

    dates, dois = (
        df
        .groupby(
            pl.col('Date').dt.year()
        )
        .agg([
            pl.col('Doi')
        ])
        .sort('Date')
    )

    dates = dates.to_numpy()
    dois = dois.to_numpy()
    
    lst_dois = [np.concatenate(dois[:i+1]) for i in range(len(dois))]
    
    subgraphs = {date: nx.subgraph(G.subgraph(dois), max(nx.weakly_connected_components(G.subgraph(dois)), key=len)) for date, dois in zip(dates, lst_dois)}

    avg_shortest_paths = {date: nx.average_shortest_path_length(subgraph) for date, subgraph in subgraphs.items()}
    
    dates = list(avg_shortest_paths.keys())
    avg_shortes_paths = list(avg_shortest_paths.values())

    fig = plt.figure(figsize=(25, 5))
    axs = fig.subplots(1, 1)

    axs.plot(dates, avg_shortes_paths)
    axs.set_xlabel('Year')
    axs.set_ylabel('Average Shortest Path Length')

    fig.patch.set_facecolor('white')
    fig.suptitle(f'{field_name.capitalize()}', fontsize=16)
    plt.xticks(rotation=45)

    # plt.savefig(f'{img_path}/avg_shortest_path_length_{field_name}.png')

    plt.show()
```

```python
cvsopts = dict(plot_height=400, plot_width=400)

def nodesplot(nodes, name=None, canvas=None, cat=None):
    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    aggregator=None if cat is None else ds.count_cat(cat)
    agg=canvas.points(nodes,'x','y',aggregator)
    return tf.spread(tf.shade(agg, cmap=["#FF3333"]), px=3, name=name)

def edgesplot(edges, name=None, canvas=None):
    canvas = ds.Canvas(**cvsopts) if canvas is None else canvas
    return tf.shade(canvas.line(edges, 'x','y', agg=ds.count()), name=name)

def graphplot(nodes, edges, name="", canvas=None, cat=None):
    if canvas is None:
        xr = nodes.x.min(), nodes.x.max()
        yr = nodes.y.min(), nodes.y.max()
        canvas = ds.Canvas(x_range=xr, y_range=yr, **cvsopts)
        
    np = nodesplot(nodes, name + " nodes", canvas, cat)
    ep = edgesplot(edges, name + " edges", canvas)
    return tf.stack(ep, np, how="over", name=name)
```
